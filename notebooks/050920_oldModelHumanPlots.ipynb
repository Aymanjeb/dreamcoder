```python
import seaborn as sns
from pythonlib.tools.pandastools import *

def filterOnlyStimWithAllModels(df, behexpt="2.4", MODELS = 
                                ["S12.10.test5", "S13.10.test5", "S12.10.test5_randomperm", "S13.10.test5_randomperm"]):
    """given a dataframe, filters so that only only keeps stim that have dat for all models.
    Assumes that only expect one datapoint in df per model (i.e, need to aggregate first)"""
    from analysis.importDrawgood import dgexpt
    teststims = dgexpt.getTestStim(behexpt, removepng=True)
    teststims_good =[]
    for stim in teststims:
        successful_models = set(df[df["stim"]==stim]["model"])
        if len([s for s in successful_models if s in MODELS])==len(MODELS):
            teststims_good.append(stim)   
    
    return df[df["stim"].isin(teststims_good)].reset_index()

def aggregOverStim(df):
    """averages over all stimuli for a given human/model
    note: will not check for you whether data are balanced"""
    return aggregGeneral(df, group=["human", "model", "human_cond", "planner"], values=["dist"], aggmethod=["mean"])


def applyFunctionToGroups(df_human, groups=["human", "human_cond", "planner"], model1="S12.10.test5", model2="S13.10.test5"):
    """computes the model2 minus model1
    df_human must already have aggrated so that each human one datapoint"""
    F = lambda x: x.loc[x["model"]==model2]["dist"].values[0] - x.loc[x["model"]==model1]["dist"].values[0]  # note, the [0] is correct, since only one value and is in a list
    Fname = model2 + "-" + model1
    df_human = df_human.groupby(groups).apply(F).reset_index(name=Fname)
    colname = Fname
    return df_human, colname


# For each human, take model and subtract random model (i.e, normalize to random parses)
def aggregOverStimAndMinusRand(df):
    """given raw df (human x stim = data), averages over all stim, then subtracts random models from acrtual mdoels"""
    df_human = aggregOverStim(df)

    df_human_minusrand1, colname1 = applyFunctionToGroups(df_human, model1="S12.10.test5", model2="S12.10.test5_randomperm")
    df_human_minusrand1 = df_human_minusrand1.rename(columns={colname1:"dist"})
    df_human_minusrand1["model"]=colname1
#     print(df_human_minusrand1)
    print(df_human_minusrand1)
    
    df_human_minusrand2, colname2 = applyFunctionToGroups(df_human, model1="S13.10.test5", model2="S13.10.test5_randomperm")
    df_human_minusrand2 = df_human_minusrand2.rename(columns={colname2:"dist"})
    df_human_minusrand2["model"]=colname2
    print(df_human_minusrand2)
    
#     df_human_minusrand = df_human_minusrand1.merge(df_human_minusrand2.loc[:, ["human", colname2]], on="human")
    df_human_minusrand = df_human_minusrand1.merge(df_human_minusrand2.loc[:, ["human", "dist", "model"]], on="human")
#     df_human_minusrand = df_human_minusrand1.merge(df_human_minusrand2, how="left"
    return df_human_minusrand, colname1, colname2


# For each human, take model and subtract random model (i.e, normalize to random parses)
def aggregOverStimAndMinusRand(df):
    """given raw df (human x stim = data), averages over all stim, then subtracts random models from acrtual mdoels"""
    df_human = aggregOverStim(df)

    # sep[arately do nroamlziastion for each model separately
    df_human_minusrand1, colname1 = applyFunctionToGroups(df_human, model2="S12.10.test5", model1="S12.10.test5_randomperm")
    df_human_minusrand2, colname2 = applyFunctionToGroups(df_human, model2="S13.10.test5", model1="S13.10.test5_randomperm")

    # merge into one dataframe
    df_human_minusrand = df_human_minusrand1.merge(df_human_minusrand2.loc[:, ["human", colname2]], on="human")
    
    # convert from wide to long form
    df_human_minusrand = df_human_minusrand.rename(columns={colname1:"dist1", colname2:"dist2"})
    df_human_minusrand = pd.wide_to_long(df_human_minusrand, stubnames="dist", i=["human"], j="model")
    df_human_minusrand = df_human_minusrand.reset_index()
    return df_human_minusrand, colname1, colname2


# 2) HUMAN = DATAPOINT, i.e, averaging over all stimuli. 
# Note, to make this a fair comaprison between models, only use stimuli that all models solved
# This is usually the "test" stimuli.

def plotPointHumanOverview(df, commonstim=True, subtractrandomparses=False, dosave=True, savesuffix="",
                           MODEL_LIST = ["S12.10.test5", "S13.10.test5", "S12.10.test5_randomperm", "S13.10.test5_randomperm"]):
    """ dat = human, averaged over all stim, and if commonstim is true, then makes
    sure only gets stimuli that all models did.
    df is human x stim (i.e., point)
    subtractrandomparses, then for each model gets difference from the "randomperm" model.
    """
    from modelPlanning import addLabel

#     MODEL_LIST = ["S12.10.test4", "S13.10.test4", "S12.10.test4_randomperm", "S13.10.test4_randomperm"]
    # MODEL_LIST = ["S12.10", "S13.10.test4", "S12.10_randomperm", "S13.10.test4_randomperm"]

    # 1) get only common stim
    if commonstim:
        dfthis_teststimonly = filterOnlyStimWithAllModels(df, 
                                             behexpt=BEHEXPT, 
                                             MODELS = MODEL_LIST)
    else:
        dfthis_teststimonly = df
        
    # 2) Aggregate by take mean over stimuli
    dfthis_teststimonly = aggregOverStim(dfthis_teststimonly)
    
    # 3) 
    if subtractrandomparses:
        dfthis_teststimonly, colname1, colname2 = aggregOverStimAndMinusRand(dfthis_teststimonly)
    
    # 3) Plots
    # === 1) 2 visualizatinos of same plot, overview
    ax1 = sns.catplot(data=dfthis_teststimonly, hue="model", x="human_cond", y="dist", dodge=True)
    ax2 = sns.catplot(data=dfthis_teststimonly, hue="model", x="human_cond", y="dist", dodge=True, kind="boxen")

    # === 2) breking out by human
    plt.figure(figsize=(20,10))
    ax3 = sns.catplot(data=dfthis_teststimonly, x = "human", row="human_cond", y="dist", hue="model", height=3, aspect=15/3)
    addLabel(ax3)
    
    if dosave:
        ax1.savefig("{}/pointhuman_teststim_overview0_{}_{}.pdf".format(SDIRPLOTS, pname, savesuffix))
        ax2.savefig("{}/pointhuman_teststim_overview1_{}_{}.pdf".format(SDIRPLOTS, pname, savesuffix))
        ax3.savefig("{}/pointhuman_teststim_eachhuman_{}_{}.pdf".format(SDIRPLOTS, pname, savesuffix))


        
        

def plotPointHumanCompareModels(df, params_planner, model2 = "S13.10.test5", model1 = "S12.10.test5", saveon=True, savesuffix=""):
    def getStatsOLS(dfthis_teststimonly, colname):
        from statsmodels.regression.linear_model import OLS
        from statsmodels.tools import add_constant

        res_dict = {}
        
        X = dfthis_teststimonly[colname].values.reshape(-1,1)
        y = dfthis_teststimonly["x4"].values.reshape(-1,1)
        X = add_constant(X, prepend=False)

        res = OLS(y, X).fit()
        res_dict["DC_vs_planner"]=res

        X = dfthis_teststimonly[colname].values.reshape(-1,1)
        y = dfthis_teststimonly["human_cond"].values.reshape(-1,1)
        X = add_constant(X, prepend=False)

        res = OLS(y, X).fit()
        res_dict["DC_vs_humancond"]=res
        
        return res_dict


    # 1) get only common stim
    dfthis_teststimonly = filterOnlyStimWithAllModels(df, 
                                             behexpt=BEHEXPT, 
                                             MODELS = [model1, model2])
    # 2) Aggregate by take mean over stimuli
    dfthis_teststimonly = aggregOverStim(dfthis_teststimonly)
    
    # 2c) for each worker get a diferences between moedls.
    dfthis_teststimonly, colname = applyFunctionToGroups(dfthis_teststimonly, model1=model1, model2=model2)

    # check that set of workers are identical from two experimetns
#     print(len(set(dfthis_teststimonly["human"])))
#     print(len(set([col for col in df_params_workers.columns if col not in ["xname", "xnum"]])))
    assert set(dfthis_teststimonly["human"]) == set([col for col in params_planner.columns if col not in ["xname", "xnum"]]), "names of workers do not perfeclty match between drawgood and dreamcoder"

    # merge workers and params
    dfthis_teststimonly = pd.merge(dfthis_teststimonly, params_planner.transpose().reset_index().rename(columns = {'workerID':'human'}), on="human")
    dfthis_teststimonly = dfthis_teststimonly.rename(columns={i:"x{}".format(i) for i in range(6)})


    # -- plot
    res_dict = getStatsOLS(dfthis_teststimonly, colname)
    for k, v in res_dict.items():
        print(k)
        print(v.summary())

    plt.figure(figsize=(20,30))
    ax = plt.subplot(3,2,1)
    r2 = res_dict["DC_vs_planner"].rsquared
    pval = res_dict["DC_vs_planner"].pvalues[0]
    plt.title("r^2={}, p={}".format(r2, pval))
    ax = sns.regplot(x=colname, y="x4", ci=95, data=dfthis_teststimonly)
    ax = sns.scatterplot(x=colname, y="x4", hue="human_cond", data=dfthis_teststimonly)

    ax = plt.subplot(3,2,2)
    ax = sns.scatterplot(x="x4", y="x5", hue="human_cond", data=dfthis_teststimonly)

    ax = plt.subplot(3,2,3)
    sns.scatterplot(x=colname, y="x4", hue="x5", data=dfthis_teststimonly, ax=ax)

    ax = plt.subplot(3,2,4)
    sns.scatterplot(x=colname, y="x4", hue="x3", data=dfthis_teststimonly, ax=ax)

    ax = plt.subplot(3,2,5)
    sns.stripplot(data=dfthis_teststimonly, x="human_cond", y=colname, color='k')
    sns.boxplot(data=dfthis_teststimonly, x="human_cond", y=colname)
    r2 = res_dict["DC_vs_humancond"].rsquared
    pval = res_dict["DC_vs_humancond"].pvalues[0]
    plt.title("r^2={}, p={}".format(r2, pval))

    ax = plt.subplot(3,2,6)
    sns.stripplot(data=dfthis_teststimonly, x="human_cond", y="x4", color='k')
    sns.boxplot(data=dfthis_teststimonly, x="human_cond", y="x4")

    
    if saveon:
        plt.savefig("{}/pointhuman_teststim_vsplanner_{}_{}.pdf".format(SDIRPLOTS, pname, savesuffix))


```

```python
# --- for each dataset, calculate 4 numbers representing model scores for the 2 model x 2 behavior

def getOneScoreForEachModelByHumcond(df, commonstim=True, subtractrandomparses=False, MODEL_LIST = ["S12.10.test5", "S13.10.test5", "S12.10.test5_randomperm", "S13.10.test5_randomperm"]):
    """ First gets dat = human, averaged over all stim, and if commonstim is true, then makes
    sure only gets stimuli that all models did.
    df is human x stim (i.e., point).
    Then gets one value for each human cond x model by averaging over all subjects
    subtractrandomparses, then for each model gets difference from the "randomperm" model.
    """

    # 1) get only common stim
    if commonstim:
        dfthis_teststimonly = filterOnlyStimWithAllModels(df, 
                                             behexpt=BEHEXPT, 
                                             MODELS = MODEL_LIST)
    else:
        dfthis_teststimonly = df
    # 2) Aggregate by take mean over stimuli
    dfthis_teststimonly = aggregOverStim(dfthis_teststimonly)
    
    if subtractrandomparses:
        assert False, "I havent thought through why I would do this..."
        dfthis_teststimonly, colname1, colname2 = aggregOverStimAndMinusRand(dfthis_teststimonly)

    return dfthis_teststimonly


# === figure out if is the same (i.e., for each row, give a label based on information about what model matches up
# with what human condition)
#     if (model=="S12.10.test5") & (human_cond==0):
#         return True
#     else:
#         return False

def applyDoesModelMatchBehavior(df):
    """# === figure out if is the same (i.e., for each row, give a label based on information about what model matches up
    # with what human condition)
    can give any df, as long as each row has model and human cond"""
    
    def F(model, human_cond):
        """ for applygint to each row - if model and human cond matched, then true"""
        modeldict = {
            "S12.10.test5":0,
            "S12.10.test5_randomperm":0,
            "S13.10.test5":1,
            "S13.10.test5_randomperm":1
        }
        if modeldict[model]==human_cond:
            return True
        else:
            return False
        
    dfout = df.merge(df.apply(lambda x: F(x["model"], x["human_cond"]), axis=1).reset_index(), left_index=True, right_index=True).rename(columns={0:"model_match_human"})
    return dfout

def applyFunctionToAllRows(df, F, newcolname="newcol"):
    """F is applied to each row. is appended to original dataframe
    F(x) must take in x, a row object"""
    return df.merge(df.apply(lambda x: F(x), axis=1).reset_index(), left_index=True, right_index=True).rename(columns={0:newcolname})


def renameModelClass(df):
    """i.e., S12.10 and S13.10 become dc; S12.10 and S13.10 become randomparse"""
    
    def F(x):
        if x["model"] in ["S12.10.test5", "S13.10.test5", "S12.10.test4", "S13.10.test4"]:
            return "dc"
        elif "randomperm" in x["model"]:
            return "randomparse"
        else:
            assert False, "model class not known"
    
    return applyFunctionToAllRows(df, F, "model_class")

def scoreModelClass(df, getoneval=False, onlykeepmatchhuman=True):
    """ for each model class, output one score for each subject, which is the score for that subject using
    the model that was trained correspodnign to the subject. e.g., for model class dc, a score for subject trained
    on stimset S12 woudl the  the score that the S12 model got. The S13 model will not contribute to this.
    generally should give a df that already has been filtered to only include test stimuli"""

    df = aggregOverStim(df)
    df = applyDoesModelMatchBehavior(df)
    df = renameModelClass(df)
    df = df.groupby(["human", "model_match_human", "model_class", "planner"])["dist"].agg("mean").reset_index()
    if onlykeepmatchhuman:
        df = df.loc[df["model_match_human"]==True, :]
    if getoneval:
        # then aggregates over all humans
        df = df.groupby(["model_class", "planner"])["dist"].agg("mean").reset_index()
        
    return df


dfthis = dfweighted_dict["motor_prob"]
dfthis = filterOnlyStimWithAllModels(dfthis)
scoreModelClass(dfthis, getoneval=True)
    
##################### STATS
def statsTtestPaired(df1, df2, varname):
    from scipy import stats
    """ttest between two variables, assume it is paired at level of human (i.e, eachhuman 1pt each var pts)"""
    df12 = pd.merge(df1, df2, on="human")
#     print(df12)
    return stats.ttest_rel(df12[f"{varname}_x"], df12[f"{varname}_y"])
    
def filterPlannerModelclass(df, planner, model_class):
    """given a dataframe, looks for columns with given planenr and model_Class
    e.g.: df1 = dftmp[(dftmp["planner"]=="equal_prob") & (dftmp["model_class"]=="randomparse")]
    """
    return df[(df["planner"]==planner) & (df["model_class"]==model_class)]

df1 = filterPlannerModelclass(dftmp, "equal_prob", "randomparse")
df2 = filterPlannerModelclass(dftmp, "equal_prob", "dc")

# df1
res = statsTtestPaired(df1, df2, "modhu_match-notmatch")


def statsSaveTtest(df, var, planner_list, mc_list, fname):
    """given dataframe with paired data (i.e., each human has one datapt per category), then this 
    gets and saves pvals in a text file (gets all paris between all combos of planner and modelclas)"""
    with open(fname, "w") as f:
        for p1 in planner_list:
            for p2 in planner_list:
                for mc1 in mc_list:
                    for mc2 in mc_list:
                        df1 = filterPlannerModelclass(df, p1, mc1)
                        df2 = filterPlannerModelclass(df, p2, mc2)
                        res = statsTtestPaired(df1, df2, var)
                        f.write(f"{p1}-{mc1} vs {p2}-{mc2} : {res}\n")

    
```
