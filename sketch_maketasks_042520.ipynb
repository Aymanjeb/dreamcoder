```python
def plotAllTasks(Tasks):
    import matplotlib.pyplot as plt
    ncol = 6
    nrow = int(np.ceil((len(Tasks)/ncol)))
    plt.figure(figsize=(ncol*2, nrow*2))
    for i, t in enumerate(Tasks):
        ax = plt.subplot(nrow, ncol, i+1)
        sketch = executeSketch(t.original)[0]
        strokes = renderPlan(sketch, return_strokes=True)
        plotOnAxes(strokes, ax=ax)
        
# Tasks = getTasks("practice_shaping", 15)
# Tasks = getTasks("practice", 15)
Tasks = getTasks("v2.1", 15)
plotAllTasks(Tasks)
```
